```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
filepath = "../data/gold/"
```

```python
# Cargar los archivos CSV
dim_plans_df = pd.read_csv(filepath + "dim_plans.csv")
dim_users_df = pd.read_csv(filepath + "dim_users.csv")
facts_df = pd.read_csv(filepath + "fact_usage.csv")
```

```python
dim_plans_df
```

```python
dim_users_df
```

```python
facts_df
```

```python
dim_plans_df.rename(columns={"plan_name": "plan"}, inplace=True)
```

```python
df_model = facts_df.merge(dim_users_df, on="user_id", how="left") \
                     .merge(dim_plans_df, on="plan", how="left")
```

```python
df_model
```

```python
df_model.info()
```

```python
# new features
```

```python
df_model["extra_minutes"] = (df_model["calls_minutes"] - df_model["minutes_included"]).clip(lower=0)
df_model["extra_messages"] = (df_model["messages_sent"] - df_model["messages_included"]).clip(lower=0)
df_model["extra_data_mb"] = ((df_model["data_used_mb"] - df_model["mb_included"]).clip(lower=0))
```

```python
df_model
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Revisar la distribución de las variables clave
cols_to_check = ["calls_minutes", "messages_sent", "data_used_mb", "total_cost", "extra_minutes", "extra_messages", "extra_data_mb"]

plt.figure(figsize=(12, 8))
for i, col in enumerate(cols_to_check, 1):
    plt.subplot(3, 3, i)
    sns.histplot(df_model[col], bins=50, kde=True)
    plt.title(f"Distribución de {col}")

plt.tight_layout()
plt.show()
```

```python
# logs / escalados
```

```python
import numpy as np

cols_log_transform = ["calls_minutes", "messages_sent", "data_used_mb", "total_cost", "extra_minutes", "extra_messages", "extra_data_mb"]

for col in cols_log_transform:
    df_model[col] = np.log1p(df_model[col])  # log1p(x) = log(1 + x) para evitar log(0)
```

```python
df_model
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 8))
for i, col in enumerate(cols_log_transform, 1):
    plt.subplot(3, 3, i)
    sns.histplot(df_model[col], bins=50, kde=True)
    plt.title(f"Distribución transformada de {col}")

plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Definir las columnas a escalar
cols_to_scale = ["calls_minutes", "messages_sent", "data_used_mb", "total_cost", 
                 "extra_minutes", "extra_messages", "extra_data_mb"]

# Inicializar el escalador
scaler = StandardScaler()

# Aplicar el escalado solo a las columnas seleccionadas
df_model[cols_to_scale] = scaler.fit_transform(df_model[cols_to_scale])
```
